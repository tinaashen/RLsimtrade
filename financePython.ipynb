    1. Federal reserves in discussion of adjusting interest rates

    2. Recent political changes between Russia and Ukraine

    3. China's possibility of banning outsourced tech product (especially applies for AAPL)
    
    4. 2019: Tech companies thrived during the pandemic 


#### 1. Federal reserves in discussion of adjusting and raising interest rates

    General rule of thumb:
- When the interest rates are raised, the stock market go down. Bonds become more attractive. When interest rates rise, it can make borrowing money for a company more expensive, which means they have less money to invest back in the company and less cash flow stability, which typically puts pressure on share prices. When interest rates fall, the inverse is true for all of the above.
- To shield from risk, investors may consider opting for shorter-term bonds or shorter life alternative assets.

```python
aapl = close.AAPL.copy().to_frame()
aapl
```

```python
aapl['firstLag'] = aapl.shift(periods=1)
```

```python
aapl['diff'] = aapl.AAPL.sub(aapl.firstLag)
aapl['% change'] = aapl.AAPL.div(aapl.firstLag).sub(1).mul(100)
aapl
```

```python
aapl['secDiff'] = aapl.AAPL.diff(periods=1)
aapl['% second change'] = aapl.AAPL.div(aapl.firstLag)
aapl
```

```python
aapl['% second change'] = aapl.AAPL.pct_change(periods = 1).mul(100)
```

```python
del aapl['secDiff']
del aapl['% change']
del aapl['firstLag']
del aapl['diff']
```

```python
aapl
```

```python
aapl.rename(columns= {"% second change": "Change"}, inplace = True)
```
